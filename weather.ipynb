```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.hist(data['Temp_C'])
plt.show()
```

```python
plt.hist(data[data['Temp_C']>0])
plt.show()
```

```python
plt.bar(data['Weather'], data['Temp_C'])
plt.show()
```

```python
weather=np.array(data['Weather'][10:20])
temperature =np.array(data['Temp_C'][10:20])
```

```python
plt.bar(weather, temperature)
plt.show()
```

```python
plt.scatter(weather, temperature)
plt.show()
```

Prikazati dva grafove (vrijeme-temperatura, temperatura-brzina vjetra)

```python
data.head(1)
```

```python
wind =np.array(data['Wind Speed'][10:20])
```

```python
plt.subplot(1,2,1)
plt.plot(weather, temperature)
plt.subplot(1,2,2)
plt.plot(wind, temperature)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(weather, temperature)
plt.xlabel('Weather')
plt.ylabel('Temperature')
plt.subplot(1,2,2)
plt.plot(wind, temperature)
plt.xlabel('Wind')
plt.ylabel('Temperature')
plt.show()
```
